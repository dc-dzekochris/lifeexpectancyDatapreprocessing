##step4: exploratory data analysis

```python
#descriptive statistics
#a. numerical
df.describe().transpose()
```

```python
#b. categorical statistics
df.describe(include='object').transpose()
```

```python
#data distribution on histograms
#they represent a uni-variant analysis where we only analyze one column
sns.histplot(data=df,x=df['Life expectancy '],kde=True)
plt.title('distribution of life expectancy')
plt.figure(figsize=(10,5))
plt.show()
```

```python
#distribution of all numerical columns
df.select_dtypes(include='number').columns
for i in df.select_dtypes(include='number').columns:
  sns.histplot(data=df,x=df[i],kde=True)
  plt.title(f'distribution of {i}')
  plt.figure(figsize=(10,5))
  plt.show()
```

```python
#outlier detection using box-plots
#unusual distribution  within a dataset column
sns.boxplot(data=df,x=df["Life expectancy "])
plt.show()
```

```python
#distribution of all numerical columns
df.select_dtypes(include='number').columns
for i in df.select_dtypes(include='number').columns:
  sns.boxplot(data=df,x=df[i])
  plt.title(f'outliers in {i}')
  plt.show()
```

```python
#correlation analysis
#range btwn 0 and 1
#numerical columns
correlations= df[df.select_dtypes(include='number').columns].corr()
plt.figure(figsize=(15,15))
sns.heatmap(correlations,annot=True)
plt.show()
```

```python
#linear relationships
#BI-VARIANT
sns.scatterplot(data=df,y='Adult Mortality',x='Life expectancy ')
plt.show()
```

```python
for i in df.select_dtypes(include='number').columns:
  sns.scatterplot(data=df,y=df[i],x='Life expectancy ')
  plt.show()
```
